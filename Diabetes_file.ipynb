```python
#Train and splitting data
```

```python
x_test, x_train, y_test, y_train = train_test_split(x,y,test_size = 0.2)
```

```python
#Classifier Algorithms
```

```python
#Decision Tree Classifier
#RandomForest Classifier
#GradientBoost Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
model_D = DecisionTreeClassifier().fit(x_train, y_train)
model_G = GradientBoostingClassifier().fit(x_train, y_train)
model_R = RandomForestClassifier().fit(x_train, y_train)
```

```python
models=[model_D, model_G, model_R]
```

```python
for i in models:
    for j in range (len(models)):
        result_j= i
        y_predict_j = result_j.predict(x_test)
        accuracy = accuracy_score(y_predict_j, y_test)
    print(f'{i}_{accuracy}')
```
